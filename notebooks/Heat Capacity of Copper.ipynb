Temperature Pulse Model
-----------------------

```python
def make_power_pulse(energy=1., start=0., duration=1., lag=.1):
    max_power = energy/duration
    def pulse(t):
        t -= start
        if t < 0 or t > duration+lag: return 0.
        if t < lag: return max_power*t/lag
        if t > duration: return max_power*(lag+duration-t)/lag
        return max_power
    return vectorize(pulse)
```

```python
test_pulse = make_power_pulse()
test_time = linspace(-.1, 1.2, 1000)
plot(test_time, test_pulse(test_time))
title('Power Pulse'); xlabel('Time [s]'); ylabel('Power [E/s]');
```

```python
def make_simple_dtemp_dt(pulse, heat_capacity=1., reservoir_temp=10., 
                     thermal_resist=lambda temp: 1.):
    def dtemp_dt(temp, t):
        dq_dt = pulse(t) - (temp-reservoir_temp)/thermal_resist(reservoir_temp)
        return dq_dt/heat_capacity
    return dtemp_dt
```

```python
from scipy.integrate import odeint
def solve_dtemp_dt(dtemp_dt, start_temp, time_range):
    return odeint(dtemp_dt, start_temp, time_range)
```

```python
test_dtemp_dt = make_simple_dtemp_dt(make_power_pulse(energy=10.), 
                                     reservoir_temp=10.,
                                     heat_capacity=1.0,
                                     thermal_resist=lambda temp: 4.0)
test_time = linspace(0, 20, 1000)
plot(test_time, solve_dtemp_dt(test_dtemp_dt, 10., test_time))
title('Copper Sample Temperature under Power Pulse')
xlabel('Time [s]'); ylabel('Temperature [K]');
```

```python
from scipy.signal import resample
def downsample_and_add_noise(time, temp, sample_rate, noise_sigma=0.1):
    num_samples = (time[-1]-time[0])*sample_rate
    temp, time = resample(temp, num_samples, t=time)
    temp = temp[:, 0] + random.normal(0, noise_sigma, (num_samples,))
    return time, temp
```

```python
test_time = linspace(0, 20, 1000)
test_temp = solve_dtemp_dt(test_dtemp_dt, 10., test_time)
test_time, test_temp = downsample_and_add_noise(test_time, test_temp, 10.)
plot(test_time, test_temp)
title('Copper Sample Temperature with Noise')
xlabel('Time [s]'); ylabel('Temperature [K]');
```
